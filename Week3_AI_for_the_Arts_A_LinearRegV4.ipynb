### 3.2.1 Linear Regression with a Single Variable

In this example, we will build a model that uses one independent variable to predict the alcohol content of the wine. Because the seaborn visualisation above showed some correlation between alcohol and proline, we will use proline to predict alcohol

####  Step 1: Import Data

```python
from sklearn import datasets
import pandas

wine_data = datasets.load_wine()
wine = pandas.DataFrame(data=wine_data["data"], columns=wine_data["feature_names"])

wine.head()
```

Once again, create a pairplot with seaborn. Pairplot of this size may take some time - be patient!

```python
import seaborn

seaborn.pairplot(wine)
```

In the following example, we will try to predict the alcohol (called dependent variable) with the value of proline. Why would proline be a reasonable choice given the above graphs?

#### Step 2: Import the train_test_split method from the package model_selection which is included in the sklearn library.

```python
from sklearn.model_selection import train_test_split

# Subsetting our data into our dependent and independent variables.
X = wine[['proline']] 
# Using double brackets allows us to use the column headings of the padas data frame
y = wine[['alcohol']]

# Split the data. This line uses the sklearn function train_test_split().
# The test_size parameter means we can train with 75% of the data, and test on 25%. 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 123)
```

#### Step 3: Import the method **LinearRegression** from the package **linear_model** of the **sklearn** library. 

```python
from sklearn.linear_model import LinearRegression

regression = LinearRegression()
```

#### Step 4: Fit the model to the traing data and calculate the score to see how well it fit the training data. The score function is provided for the Linear Regression algorithm by the sklearn library.

```python
regression.fit(X_train, y_train)

print(regression.score(X_train, y_train))
```

Not too good! Only approx. 44.9% of the **alcohol** value is explained by the **proline** value in the training data. But how does it do on the test data?

#### Step 5: Make prediction of the total_phenols with your trained linear regression model. 

```python
y_pred = regression.predict(X_test)
```

#### Step 6: Import function `r2_score` from the package metrics of the sklearn library to measure the performance of the predictions `y_pred` above against the actual labels `y_test`, using the R-squared score (standard measure to use in linear regression). 

```python
from sklearn.metrics import r2_score

print(r2_score(y_test, y_pred))
```

The performance is far below that on the training data. However, we were only using proline values in this model. Let's see what happens when we use multiple variables to predict alcohol - in fact, all the values other than alcohol will be used in the next section.

### 3.2.2 Linear Regression with Multiple Variables

Using only proline, the prediction of the alcohol content was rather disappointing. Let us see what happens when we use all the variables other than alcohol.

#### Step 1: Save the alcohol in variable `y_mult` (traditionally the variable y represents the target value - the value you are trying to predict - but here we want to distinguish it from our previous task).

```python
y_mult = wine[["alcohol"]]
```

#### Step 2: Save the rest of the columns in variable X_mult (traditionally X used for the independent variables with which you are trying to predict the target value, but here we want to distinguish it from the previous task).

To do this you can use the `drop()` function of a pandas data frame.

```python
X_mult = wine.drop(["alcohol"], axis=1) 
# the axis argument tells the function that you are dropping a column
```

#### Step 3: Split the data into training and test data. 

```python
# Split the data into training and test data. 
#This uses the sklearn function train_test_split which we need to import form sklearn.model_selection
# The test_size parameter means we can train with 75% of the data, and test on 25%. 

from sklearn.model_selection import train_test_split

X_train_mult, X_test_mult, y_train_mult, y_test_mult = train_test_split(X_mult, y_mult, test_size = 0.25, random_state = 1)
```

#### Step 4: Get a linear regression model to train on the data - different name for the variable from before so that we can compare them later.

```python
from sklearn.linear_model import LinearRegression
# you don't need to import again if you've done it in a previous exercise
# importing it here again so that this section is self contained.

multiple_regression = LinearRegression()
```

#### Step 5: Fit the model to your training data.

```python
multiple_regression.fit(X_train_mult, y_train_mult)

print(multiple_regression.score(X_train_mult, y_train_mult))
```

Yes! It does seem to help when you use more variables. But we have not tested it yet with test data.

#### Step 6: Make predictions on your test data.

```python
y_pred_mult = multiple_regression.predict(X_test_mult)
```

#### Step 7: Use R-Squared to compare your predictions against the actual values y_test.

```python
from sklearn.metrics import r2_score

print(r2_score(y_test_mult, y_pred_mult))
```

Well, what happened? The performance is much better than when we used only a single variable for prediction. This does not always happen, for example if the model overfits the data (see Lecture this week).

So far, we have used **R-Squared** as a means of carrying out evaluation. However, sometimes you want to visualise data to get a feel for how your approaches are doing. This also allows you to cross check whether your scores are actually doing what they are supposed to! We will do that next to compare your models `regression` to `multiple_regression`.

### 3.2.3 Visualising Performance

#### Step 8: Visualise the predicted values against the actual values for both the single variable regression and multiple variable regression. 

```python
from matplotlib import pyplot
minV = 11 ## minimum value of alcohol in the dataset
maxV = 15 ## maximum value of alcohol in the dataset

# Plot the predictions
# Build a scatterplot of predictions against actual values of alcohol
pyplot.figure()
pyplot.scatter(y_test, y_pred, color='green', marker='D', s=10) #single variable regression 
pyplot.scatter(y_test_mult, y_pred_mult, color='red', marker='D', s=10)# multiple variable regression
pyplot.plot([minV,maxV], [minV,maxV], color="blue", linewidth=3) 
# line for perfect correlation between prediction and actual

# Label it nicely
pyplot.title('Linear regression: single variable (green) and multiple variables (red)')
pyplot.xlabel('actual alcohol')
pyplot.ylabel('predicted alcohol')
pyplot.grid(True)
```
